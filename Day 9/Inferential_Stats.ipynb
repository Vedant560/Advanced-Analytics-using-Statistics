```python
n=300
x=1800
stddev=70
cl=0.95

```

```python
low_est=x-t*stddev/n**0.5     # normal Distrribution
upp_est=x+t*stddev/n**0.5
int_size=upp_est-low_est
```

```python
low_estz=x-z*stddev/n**0.5     # t distribution (Sudent's t)
upp_estz=x+z*stddev/n**0.5
int_size1=upp_estz-low_estz
```
