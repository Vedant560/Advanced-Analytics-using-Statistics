# Group By Advanced

```python
df=pd.read_excel("CDAC_DataBook.xlsx",sheet_name="ERPData")
```

```python
df.head()
```

```python
np.random.seed(42)
x1=np.random.randint(10,80,50)
df["Value"]=x1
```

```python
df.head()
```

```python
#grouping can be done on Categorical Data
grp1=df.groupby("MaterialID")# Grouping done based on materialID
```

```python
grp1.groups #Displaying the Groups
```

```python
grp1.get_group("AXCP-78")
```

```python
grp1.Quantity.agg(["sum","mean","size","var"])
```

```python
s1=grp1.Quantity.agg(["size"])
d1=pd.DataFrame(s1)
d1.loc["SDRT-67"]
```

```python
d1
```

```python
np.where(d1.index=="SDRT-67")
```

```python
d1.iloc[np.where(d1.index=="AXCP-78")]
```

```python
d2=grp1.Quantity.agg(["sum","mean","size","var"])
```

```python
d2
```

```python
d2.loc["SDRT-67","sum"]
```

# Reshaping

```python
df=pd.read_excel("CDAC_DataBook.xlsx",sheet_name="Health")
```

```python
df
```

```python
df1=df.melt(id_vars=["Country","Disease"],var_name="Year",value_name="Cases") # Coverting the df int df1 
#making the df1 tall structure
df1
```

```python
df2=df1.pivot_table(values="Cases",index=["Country","Year"],columns="Disease")
df2
```

```python
df2.shape
```

```python
df2.index
```

```python
df2.columns
```

```python
df3=df1.pivot_table(values="Cases",index=["Country","Year"],columns="Disease").reset_index().rename_axis(columns=None)
#Converting the labels(indices) into columns and renamed it
df3
```

```python
df3.shape
```

```python
df3.index
```

```python
df3.columns
```
