# 12.One Sample Poisson Test

```python
# A good class would ask alteast 20 questions everyday

# Ho: n>=20
# Ha: n<20


questions=[14,19,23,18,23,16,25,20,22,18,16,14]
np.mean(questions)
```

```python
test_poisson(sum(questions),len(questions),20,method="score",alternative="smaller")
```

# 6.One Sample Variance Test

```python
df=pd.read_excel("CDAC_DataBook.xlsx",sheet_name="faithful")
```

```python
errupt=df.eruptions
```

```python
np.var(errupt,ddof=1)
```

```python
# Ho: var(eerrupt)<1.1
# Ha: var(errupt)>=1.1 -------> Right Tail Test

# TS= (n-1)*variance/(claim value)

```

```python
ts=(len(errupt)-1)*np.var(errupt,ddof=1)/1.1   # Test Statistics
ts
```

```python
p_value=1-chi2.cdf(ts,len(errupt)-1)
p_value
```

```python
np.var(errupt,ddof=1)/1.1
```

# 7. Two Sample Variance Test


```python
samp1=errupt[30:70]
samp2=errupt[130:180]
```

```python
samp_ratio=np.var(samp1,ddof=1)/np.var(samp2,ddof=1)
samp_ratio
```

```python
# Ho: var(samp1)/var(samp2)>=1.4
# Ha: var(samp1)/var(samp2)<1.4
```

```python
ts=samp_ratio/1.4
ts
```

```python
f.cdf(ts,len(samp1)-1,len(samp2)-1)
```
