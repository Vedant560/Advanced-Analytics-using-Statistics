Linear Algebra


```python
from numpy import linalg
```

```python
A = np.array([
    [1,2,3],
    [0,1,4],
    [5,6,0]
])
```

```python
np.linalg.det(A)
```

```python
np.transpose(A)
```

```python
np.linalg.inv(A)
```

```python
np.invert(A)
```

```python
np.linalg.eigvals(A)
```

```python
np.sum(A,axis=0)
```

```python
np.sum(A,axis=1)
```

## Challenge 3

Take two arrays and:

Find common elements
Find unique elements
Merge them
Sort the final array

```python
a1 = np.array([1,2,3,4,5,6])
a2 = np.array([4,5,6,7,8,9])
```

```python
np.intersect1d(a1,a2)
```

```python
np.setxor1d(a1,a2)
```

```python
merged=np.union1d(a1,a2)
merged
```

```python
np.sort(merged)
```
