# Set operation

```python
a1=np.array([3,7,6,1,8,9])
a2=np.array([2,4,5,8,10,3,])
```

```python
# "and" means Intersection  (Common in both sets)
# "or"  means union (all elements in both sets without duplicates)
```

```python
np.intersect1d(a1,a2) # result would be same if the sets are intechanged
```

```python
np.union1d(a1,a2) #all the values in set a1 and a2
```

```python
np.isin(a1,a2) #output will be a logical array
```

```python
np.isin(a2,a1)
```

```python
np.setdiff1d(a1,a2) # elements that are unique to a1 (only present in a1 but not in a2)
```

```python
np.setdiff1d(a2,a1)  # elements that are unique to a1 (only present in a2 but not in a1)
```

```python
np.setxor1d(a1,a2) #union of unique elemnts either in a1 or in a2  (elements only in a1 or a2)
```

```python
np.where(np.isin(a1,a2)) # indices of common values 
```

**DOT & CROSS PRODUCT**

```python
x1=np.array([2,3,-4])
x2=np.array([3,-2,1])

```

```python
np.dot(x1,x2) # Dot product of 2 vectors
```

```python
from numpy import linalg #linear Algebra
```

```python
x3=np.array([[1,2,3],[0,1,2],[1,1,0]])
```

```python
np.linalg.det(x3) #Determinant
```

```python
np.cross(x1,x2) # cross product of 2 vectors
```
