```python
import numpy as np

# 1. Create a 5x5 array with random float values between 0 and 100
np.random.seed(42)  # Setting seed for reproducible results
random_array = np.random.rand(5, 5) * 100
print("Original 5x5 Random Array:")
print(random_array)
print("-" * 50)

# 2. Find the min and max values for each row (axis=1)
# we use keepdims=True to maintain the (5, 1) shape for broadcasting
row_mins = random_array.min(axis=1, keepdims=True)
row_maxs = random_array.max(axis=1, keepdims=True)

# 3. Apply the Min-Max formula row-wise
normalized_row_wise = (random_array - row_mins) / (row_maxs - row_mins)

print("Row-wise Normalized Array (0 to 1):")
print(normalized_row_wise)
```

```python
import numpy as np

# 1. Create a 5x5 array with random float values between 0 and 100
np.random.seed(52)  # Setting seed for reproducible results
random_array = np.random.rand(5, 5) * 100
print("Original 5x5 Random Array:")
print(random_array)
print("-" * 50)

# 2. Find the min and max values for each row (axis=1)
# we use keepdims=True to maintain the (5, 1) shape for broadcasting
row_mins = random_array.min(axis=0, keepdims=True)
row_maxs = random_array.max(axis=0, keepdims=True)

# 3. Apply the Min-Max formula row-wise
normalized_row_wise = (random_array - row_mins) / (row_maxs - row_mins)

print("Column-wise Normalized Array (0 to 1):")
print(normalized_row_wise)
```
