```python
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    "PAN": ["ABCDE1234F", "PQRSV6789Z", "LMNOP1234Q"],
    "Surname": ["Eshwar", "Verma", "Patel"]
})

# Regex pattern for PAN format
pattern = r'^[A-Z]{5}[0-9]{4}[A-Z]{1}$'

# Step 1: Basic format validation
df["format_valid"] = df["PAN"].str.match(pattern)

# Step 2: Extract 5th character of PAN
df["fifth_char"] = df["PAN"].str[4]

# Step 3: Extract first letter of surname (uppercase)
df["surname_initial"] = df["Surname"].str[0].str.upper()

# Step 4: Check if both match
df["name_match"] = df["fifth_char"] == df["surname_initial"]

# Final validity: both format and name match must be True
df["is_valid"] = df["format_valid"] & df["name_match"]

print(df)
```
