```python
import numpy as np
import pandas as pd
```

# Group BY 

```python
df=pd.read_excel("CDAC_DataBook.xlsx",sheet_name="ERPData") #Importing DataSet
```

```python
grp1=df.groupby("MaterialID") # Creating Gropus Based on Material ID
```

```python
grp1.groups #Displaying the Gropus that are created
```

```python
grp1.get_group("SDRT-67").reset_index(drop=True) #Displayiing group "SDRT-67
```

```python
grp1.Quantity.agg("sum") #Finding Sum of the quantity based on each Material ID
```

```python
grp2=df.groupby(["Location","MaterialID"]) # Creating Group based on 2 column
```

```python
grp2.groups
```

```python
grp2.Quantity.agg("sum")
```

```python
newdf=grp2.agg(["sum","mean","size"]) #Finding the sum , Size , mean based on the groups
newdf
```

```python
newdf.loc[["MWH-1","MWH-2","MWH-3"]] #Extracting the info of location and Material Id based on Quantity
```

# Sorting on DataFrames

```python
df_asc=df.sort_values("Quantity") #Ascending Order
df_asc.head()
```

```python
df_desc=df.sort_values("Quantity",ascending=False) #Descending Order
df_desc.head()
```

```python
df_sort_cols=df.sort_values(["MaterialID","Location","Quantity",]) #Sorting has been done based on the multiple cols in order of the column given
df_sort_cols.head()
```

```python
df_sort_cols=df.sort_values(["MaterialID","Location","Quantity",],ascending=[True,False,True])
df_sort_cols.head() #Same as above but MaterialID, Quantity is in Ascending order and location is in Descending order
```

# Merging the DataFrames

```python
pune=pd.read_excel("CDAC_DataBook.xlsx" ,sheet_name="Pune")
mumbai=pd.read_excel("CDAC_DataBook.xlsx" ,sheet_name="Mumbai")
```

```python
pune
```

```python
mumbai
```

```python
pd.merge(pune,mumbai) # this will show empty dataframe because there is no entire common row in both the dataframe
#label and index may not be the same
```

```python
pd.merge(pune,mumbai,on="Subject" ) # merging the Pune and Mumbai based on Subject  
#common subject between Pune , Mumbai      ## By default how= inner
```

```python
pd.merge(pune,mumbai,on=["Subject","Grade"],how="inner") #this will give all common cols in both tables
```

```python
pd.merge(pune,mumbai,on="Subject",how="outer")
```

```python
pd.merge(pune,mumbai,on="Subject",how="left")
```

```python
pd.merge(pune,mumbai,on="Subject",how="right")
```
