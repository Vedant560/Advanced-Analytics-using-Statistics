**2D Array**

```python
x1 = (np.arange(48)).reshape(6,8)   # creates a 2d array of 6 rows and 8 columns
x1
```

```python
x2 = np.resize(x1,(12,4))   #  resizes the array x1 which was having 6 rows and 8 cols to an array with 12 rows and 4 cols
x2
```

```python
x3 = np.split(x2,[1,3,5])	# splits the array at the indices mentioned. The resultant arrays 
# are different items in a list. So basically we get list of arrays. By defaults axis=0 so splitting is done horizontally
x3

x4=np.split(x2,[2,6,9])
x4
```

```python
x4 = np.split(x1,[2,3,5], axis=1)    # splitting happens vertically
x4
```

```python
b=np.array([2,4,1,7,6,5,9,8,2,3,1,5])
c = sorted(b)          #sorts the array in increasing order
c = sorted(b)[::-1]    # sorts the array in the decreasing order

np.sort(b)     # also PRINTS the array in the increasing order, original array is unchanged
```

```python
b=b.reshape(3,4)   # reshapes b as a 2d array with 3 rows and 4 columns
 
b
```

```python
np.sort(b,axis=0)   # row wise values will be changed. Column values are only rearranged
```

```python
np.sort(b,axis=1)   # column wise values will be changed. Row values are only rearranged
```

```python
c=np.where(b>=4)
c      #ROWS                        #COLS 
```

```python
b[c]
```

**SORT AND ARGSORT**

```python
a1=np.array([4,7,6,1,5,9,7,8,2])
```

```python
np.sort(a1) # returns value from lowest to highest
```

```python
np.argsort(a1) # returns indices of the lowest to highest value
```

```python
a1[np.argsort(a1)] # same as np.sort(a1)
```

**UNIQUE & OCCURANCE**

```python
a1=np.array([6,5,4,8,7,6,2,2,5,7,1,8,1,2,5,6,9,4,7,5,6,2,0,8,8,9,2,5])
len(a1)
```

```python
np.unique(a1) 
```

```python
np.unique(a1, return_counts=True) # Returns value and number of times that value appear
```

```python
np.unique(a1, return_index=True)   #Reurns value and first time that value occurring for the first time
```
