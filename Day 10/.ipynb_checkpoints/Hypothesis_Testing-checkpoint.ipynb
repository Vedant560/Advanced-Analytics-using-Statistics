# 1 Sample z test / ttest

```python
erupt=df["eruptions"]
```

```python
xbar=np.mean(erupt)
xbar
```

```python
n=len(erupt)
n
```

```python
sd=np.std(erupt,ddof=1)
sd
```

```python
ssw.ztest(erupt,value=3.7,alternative="smaller") # output will (test_statistics , p-value)
```

```python
# calculating the Test Statistics
ts=(xbar-3.7)/(sd/n**0.5)
ts
```

```python
zc=norm.ppf(0.05)
zc
```

Ho: u<=3.53

Ha: u>3.53

```python
ts=(xbar-3.53)/(sd/n**0.5)   # test statistics calculated for 3.53
ts                         
```

```python
norm.cdf(ts) # calculated p value for u=3.53
```

```python
ssw.ztest(erupt,value=3.53,alternative="smaller")
```

```python
df=pd.read_excel("CDAC_DataBook.xlsx",sheet_name="faithful")
```

```python
df.head()
```

```python
df1=df.iloc[20:60]
df2=df.iloc[120:170]
```

```python
erupt1=df1.eruptions
erupt2=df2.eruptions
```

```python
np.mean(erupt1),np.mean(erupt2)
```

```python
#mean of Errupt 2 is atleast 0.12 more than measn of errupt1

# Ho: mean(E2)-mean(E1)>=0.12
# Ha:  mean(E2)-mean(E1)<0.12
```

```python
ssw.ztest(erupt2,erupt1,value=0.12,alternative="smaller")
```

```python
ssw.ztest(erupt2,erupt1,value=0,alternative="two-sided")
```
