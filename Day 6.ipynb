# Probability Distribution Function

## Binomial Distribution

1. binom.pmf (pmf- probability mass function): used to find out the probablity of exact number
2. binom.cmf (cmf- cumulative distribution function): used to find out the probablity of upto a particular number
3. binom.sf  (sf-)

```python
'''
Chances of rain on any given day in June in Mumbai is 0.7
If I am in Mumbai for 15 days, what is the probability that
a. It rains for exactly 10 days
b. It rains for maximum 10 days
c. It rains for minimum 10 days
'''
```

```python
binom.pmf(10,15,0.7) #Probability of exact 
```

```python
binom.cdf(10,15,0.7)  #Probabiltiy of max 10 days of Rain
```

```python
1- binom.cdf(10,15,0.7) 
```

```python
1-binom.cdf(9,15,0.7) #Probability of min 10 days
```

## Possion Distribution

```python
'''Avg Number of occurances is 15.What is the Probbility of

a.exxactly 12 incidents  taking place today
b. upto 12 incidents take place
c. minimum 12 incidents take place

'''
```

```python
poisson.pmf(12,15) #a
```

```python
poisson.cdf(12,15) #b
```

```python
1- poisson.cdf(11,15)
```

```python
poisson.sf(11,15)
```
